```python
# importing libraries
import numpy as np
import pandas as pd
```

```python
# TASK 1
# Create a dummy dataframe (or load one) with an index of daily dates for the year 2023.
# Generate random "Sales" data for each day.
# Resample this daily data to find the Total (Sum) Sales per Month.

# Create a dummy dataframe with daily dates for the year 2023
date_range = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')

# Random sales data between 100 and 500
sales_data = np.random.randint(100, 500 + 1, size=len(date_range))  

df = pd.DataFrame({'Date': date_range, 'Sales': sales_data})

# Set 'Date' as the index
df.set_index('Date', inplace=True)  
df.head() 
```

```python
# Resample the data to find total sales per month
df_monthly_sales = df.resample("M").sum() 
df_monthly_sales
```

```python
# TASK 2
# Create a small dataframe with columns: ['Date', 'Region', 'Sales']. 
# Populate it with data for 3 regions (North, South, East) over 5 different dates.
# Use pivot() to reshape this data so that:
# -- Date is the Index.
# -- Region names are the Columns (North, South, East).
# -- Sales are the Values.

data = pd.DataFrame({"Date": pd.date_range(start='2023-01-01', periods=5, freq='D'),
                     "Region": ['North', 'South', 'East', 'North', 'South'],
                     "Sales": [200, 150, 300, 250, 100]})
data
```

```python
data_pivot = data.pivot(index='Date', columns='Region', values='Sales')
data_pivot
```

```python
# TASK 3
# import the "tips" dataset
tips = pd.read_csv("./data/tips.csv")

tips.head()
```

```python
# create a Pivot Table that shows:
# Rows: Day of the Week.
# Columns: Time (Lunch vs. Dinner).
# Values: Total Bill.
# Aggregation: Calculate the Average (mean) bill size.
# Margins: Include the "All" row/column totals.

pivot_table = tips.pivot_table(index="day", columns="time", values="total_bill", aggfunc="mean", margins=True)
pivot_table
```
